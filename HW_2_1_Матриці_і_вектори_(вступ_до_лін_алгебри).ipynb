
### Завдання 3: Множення матриць
1. Задані дві матриці `A` і `B`розмірностями 3x2 та 2х3. Виконайте їхнє множення, щоб отримати матрицю 2x2. Виведіть результуючу матрицю `result_matrix` на екран.

2. Заданий вектор `C` розмірністю 2х1. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на вектор `C` і який результат буде. А потім виконайте множення і виведіть результат на екран.

3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
result_matrix = np.dot(A, B)
print(f'Помножені матриці А та В:\n{result_matrix}\n')
print("Shape result_matrix:", result_matrix.shape, "\nShape C:", C.shape)
print('Множення result_matrix @ C неможливе так як 3 != 2 (стовпці result × рядки C)')
print("\nShape D:", D.shape)
print("Множення result_matrix на D також неможливе з тієї ж причини")
```

```python
print('Звичайне множення A × B дає 3×3, тому для множення на C (2×1) та D (2×2) використано транспонування обох матриць.')
print('Але математично рішення буде не вірним.\n')
result_matrix_2 = np.dot(A.T, B.T)
print(f'Помножені транспоновані матриці А та В:\n{result_matrix_2}\n')
result_с = np.dot(result_matrix_2, C)
print(f'Матриця помножена на вектор С:\n{result_с}\n')
result_в = np.dot(result_matrix_2, D)
print(f'Матриця помножена на матрицю D:\n{result_в}')
```
