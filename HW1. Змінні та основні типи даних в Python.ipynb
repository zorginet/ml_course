**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
var1 = 15
var2 = 4
print("Сума:", var1 + var2) # Сума
print("Різниця:", var1 - var2) # Різниця
print("Добуток:", var1 * var2) # Добуток
print("Ділення:", var1 / var2) # Ділення
print("ілочисельне ділення:", var1 // var2) # Цілочисельне ділення
print("Залишок від ділення:", var1 % var2) # Залишок від ділення
print("Піднесення до степеня:", var1 ** var2) # Піднесення до степеня
```
