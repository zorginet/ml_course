```python
median_purchases_python = (
    sorted(sum_purchases)[len(sum_purchases)//2-1] 
    + sorted(sum_purchases)[len(sum_purchases)//2]) / 2

print(f'Медіана Python: {median_purchases_python}')

median_purchases_numpy = np.median(sum_purchases)

print(f'Медіана NumPy: {median_purchases_numpy}')


```

**Завдання 5**. Ми провели дослідження віку студентів-першокурсників. Виявилось, що
- 10 студентів мають вік 16 років
- 25 студентів - 17 років
- 20 студентів - 18 років
- 2 студенти - 20 років

Обчислість середнє значення для цієї вибірки будь-яким способом - на чистому Python або з numpy - та виведіть результат на екран.

```python
ages = {16: 10,
        17: 25,
        18: 20,
        20: 2}
mean_ages = sum([key * value for key, value in ages.items()]) / sum(ages.values())

print(f'Середній вік Python: {round(mean_ages, 2)}')

import numpy as np

mean_ages_numpy = np.average([16, 17, 18, 20], weights=[10, 25, 20, 2])

print(f'Середній вік NumPy: {mean_ages_numpy.round(2)}')
```

Виведіть результат на екран. Проаналізуйте результат.

Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
corr_numpy = np.corrcoef(sum_purchases, visits)
print(f'Kоефіцієнт кореляції NumPy: \n \n {corr_numpy.round(3)}')
```

```python
import math

mean_visits = sum(visits) / len(visits)

var_visits = sum((i - mean_visits)**2 for i in visits) / (len(visits) - 1)

sqrt_visits_python = math.sqrt(var_visits)

purchases_visits = [(sum_purchases[i] - mean_purchases_python) 
                * (visits[i] - mean_visits) for i in range(len(sum_purchases))]

cov_purchases_visits = sum(purchases_visits) / (len(sum_purchases) - 1)

cov_purchases_visits

corr_python = cov_purchases_visits / (sqrt_python * sqrt_visits_python)

print(f'Kоефіцієнт кореляції Python: {corr_python:.3f}')
```

```python
import matplotlib.pyplot as plt

plt.scatter(sum_purchases, visits)
plt.title('Scatter Plot of Purchases vs. Visits')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

Коефіцієнти кореляції, обчислені вручну та за допомогою NumPy, співпадають.
Значення кореляції не є високим, що свідчить про слабку лінійну залежність між сумою покупок та кількістю відвідувань.
На графіку розсіювання помітні викиди, які можуть впливати на значення коефіцієнта Пірсона, тому перед аналізом варто перевіряти дані на наявність таких значень.

**Завдання 7**. В попередньому завданні ми помітили, що викиди нам заважають отримати "правдиве" значення кореляції. Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
import numpy as np
mean_purchases = np.mean(sum_purchases)
std_purchases = np.std(sum_purchases, ddof=1)
threshold_purchases = mean_purchases + 2 * std_purchases
new_purchases = []
new_visits = []
for i in range(len(sum_purchases)):
    if sum_purchases[i] <= threshold_purchases:
        new_purchases.append(sum_purchases[i])
        new_visits.append(visits[i])
print(new_purchases)
print(new_visits)

corr_new = np.corrcoef(new_purchases, new_visits)
print(f'Kоефіцієнт кореляції New: \n \n {corr_new.round(3)}')
```

```python
import matplotlib.pyplot as plt

plt.scatter(new_purchases, new_visits)
plt.title('Scatter Plot of Purchases vs. Visits')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

Після вилучення викидів за сумою покупок коефіцієнт кореляції Пірсона істотно не змінився.
Діаграма розсіювання показує значний розкид точок, що свідчить про відсутність сильної лінійної залежності між кількістю візитів і сумою покупок.
Отже, кількість візитів не є хорошим лінійним предиктором суми покупок.
