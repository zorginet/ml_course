### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)

plt.figure(figsize=(10, 5))
plt.plot(x, f, label= 'f = 10 * x - 5', color='blue')
plt.plot(x, df, label= "f'(x) = 10", color='green')

plt.xlabel('Години наічання')
plt.ylabel('Оцінка')
plt.legend()
plt.grid(True)
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_sin = sp.sin(x)
df_sin = sp.diff(f_sin, x)
f_sin_np = sp.lambdify(x, f_sin, 'numpy')
df_sin_np = sp.lambdify(x, df_sin, 'numpy')
x_vals = np.linspace(-2*np.pi, 2*np.pi, 200)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_sin_np(x_vals), label='sin(x)')
plt.plot(x_vals, df_sin_np(x_vals), label="cos(x)")
plt.legend()
plt.grid(True)
plt.title("f(x) = sin(x) та ї похідна")
plt.show()
```

```python
x = sp.symbols('x')
f_log = sp.log(x)
df_log = sp.diff(f_log, x)
f_log_np = sp.lambdify(x, f_log, 'numpy')
df_log_np = sp.lambdify(x, df_log, 'numpy')
x_log = np.linspace(0.1, 5, 200)

plt.figure(figsize=(10, 5))
plt.plot(x_log, f_log_np(x_log), label='ln(x)')
plt.plot(x_log, df_log_np(x_log), label='1/x')
plt.legend()
plt.grid(True)
plt.title("f(x) = ln(x) та ї похідна")
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x)
f_np = sp.lambdify(x, f, 'numpy')
df_np = sp.lambdify(x, df, 'numpy')
x_vals = np.linspace(0, 30, 100)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_np(x_vals), label='Загальні витрати f(x)', color='blue')
plt.plot(x_vals, df_np(x_vals), label="Похідна f'(x)", color='red')
plt.xlabel('Кількість вироблених одиниць')
plt.ylabel('Значення')
plt.title('Загальні витрати та їх похідна')
plt.legend()
plt.grid(True)
plt.show()
```
